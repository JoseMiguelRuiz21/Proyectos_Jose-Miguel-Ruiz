### 2.2 Renombrar columnas

```python
# COMPLETAR PASO 2.3: Limpiar datos numéricos del dataset eco
print("Limpiando datos numéricos de eco...")

# Limpiar city_gdp_capita: eliminar separadores de miles y convertir comas a puntos
eco['city_gdp_capita'] = eco['city_gdp_capita'].str.replace('.', '', regex=False)  # Eliminar separadores de miles
eco['city_gdp_capita'] = eco['city_gdp_capita'].str.replace(',', '.', regex=False)  # Cambiar comas por puntos
eco['city_gdp_capita'] = pd.to_numeric(eco['city_gdp_capita'], errors='coerce')  # Convertir a numérico

# Limpiar unemployment_pct: eliminar % y convertir comas a puntos
eco['unemployment_pct'] = eco['unemployment_pct'].str.replace('%', '', regex=False)  # Eliminar %
eco['unemployment_pct'] = eco['unemployment_pct'].str.replace(',', '.', regex=False)  # Cambiar comas por puntos
eco['unemployment_pct'] = pd.to_numeric(eco['unemployment_pct'], errors='coerce')  # Convertir a numérico

# Limpiar population_m: convertir comas a puntos
eco['population_m'] = eco['population_m'].str.replace(',', '.', regex=False)  # Cambiar comas por puntos
eco['population_m'] = pd.to_numeric(eco['population_m'], errors='coerce')  # Convertir a numérico

# Crear columna population (población total)
eco['population'] = eco['population_m'] * 1000000

# Verificar que la limpieza funcionó
print("Verificación después de la limpieza:")
print("city_gdp_capita tipo:", eco['city_gdp_capita'].dtype)
print("unemployment_pct tipo:", eco['unemployment_pct'].dtype)
print("population_m tipo:", eco['population_m'].dtype)
print("Primeros 3 valores de city_gdp_capita:")
print(eco['city_gdp_capita'].head(3))
```

### 3.1 Extraer columna año y filtrar 2024

**🎯Objetivo**
Identificar el año de cada registro y mantener solo los registros del 2024.

**Intrucciones**

- Como el DataFrame `traffic` no tiene una columna de año, utiliza el atributo `.dt.year` sobre su columna de fecha para crear una nueva columna llamada `year`.

- Utiliza `.copy()` para crear dos nuevos DataFrames (`traffic_2024` y `eco_2024`) para evitar modificar el dataset original.

```python

# Paso 1: Convertir a datetime
traffic['updatetime_utc'] = pd.to_datetime(traffic['updatetime_utc'], errors='coerce')

# Paso 2: Extraer el año
traffic['year'] = traffic['updatetime_utc'].dt.year
```

```python
# Filtra los registros del año 2024
# traffic_2024 = traffic[...].copy()
# eco_2024 = ...
traffic_2024 = traffic[traffic['year'] == 2024].copy()
eco_2024 = eco[eco['year'] == 2024].copy()



# Revisar dataframes nuevos
display(traffic_2024.head())
display(eco_2024.head())
```

### 6.1 Visualizar relaciones entre economía y tráfico

**🎯Objetivo:**
Analizar visualmente la distribución y la relación entre indicadores de tráfico y economía en 2024, para identificar posibles patrones o tendencias generales entre ambas variables.

**Instrucciones**
- Usa las librerías `seaborn` y `matplotlib.pyplot` para generar los gráficos.
- Visualiza la distribución del **tráfico** (`jams_delay`) mediante:
    - **Boxplot** → para observar la media, mediana y detectar valores atípicos.
- Visualiza la distribución de la **economía** (`city_gdp_capita`) mediante:
    - **Histograma** → para analizar la forma de la distribución y el valor promedio del PIB per cápita.
- Finalmente, **compara ambas variables**, para observar si existe alguna relación entre ellas, haciendo un solo gráfico de barras donde aparezcan ambos indicadores.
- Recuerda agregar título y etiquetas a los ejes de tus gráficos.

**Tip:** Dentro de los parentesis del boxplot, agrega `showmeans=True` para ver la media en el gráfico.

```python

# --- 1. BOXPLOT: Distribución de los minutos de congestión (jams_delay) ---
plt.figure(figsize=(10, 5))
sns.boxplot(x=merged['jamsdelay'], color='skyblue', showmeans=True)
mean_delay = merged['jamsdelay'].mean()
plt.title(f'Distribución de Retrasos por Tráfico (2024)\nPromedio: {mean_delay:.2f} min')
plt.xlabel('Retraso promedio (minutos)')
plt.show()
```

```python
# Crear histograma para ver la distribución de la economía (city_gdp_capita)
# 1. Histograma de la Economía (PIB per cápita)
plt.figure(figsize=(10,6))
sns.histplot(merged['city_gdp_capita'], kde=True, color='green')
plt.title('Distribución del PIB per cápita en las ciudades (2024)')
plt.xlabel('PIB per cápita (USD)')
plt.ylabel('Frecuencia (Número de ciudades)')
plt.show()
```

```python

# --- 3. GRÁFICO DE BARRAS COMPARATIVO: Tráfico vs Economía ---
# Usamos un gráfico con dos ejes Y porque el PIB tiene números mucho más grandes que los minutos de tráfico
fig, ax1 = plt.subplots(figsize=(15, 7))

# Barras para el tráfico
merged.plot(kind='bar', x='city', y='jamsdelay', ax=ax1, color='orange', label='Retraso (min)')
ax1.set_ylabel('Retraso por Tráfico (minutos)')
ax1.set_xlabel('Ciudad')

# Línea para el PIB (segundo eje Y)
ax2 = ax1.twinx()
merged.plot(kind='line', x='city', y='city_gdp_capita', ax=ax2, color='blue', marker='o', label='PIB per cápita')
ax2.set_ylabel('PIB per cápita (USD)')

plt.title('Comparativa: Retraso por Tráfico vs PIB por Ciudad (2024)')
ax1.set_xticklabels(merged['city'], rotation=90) # Rotamos nombres para que se lean bien
plt.show()
```

**Tip:** Antes del `plt.show()` agrega el código `plt.xticks(rotation=90)` para rotar las etiquetas del eje X en 90 grados.
